```python
import pyttsx3                 #version 2.6
import datetime
import speech_recognition as sr
import wikipedia
import webbrowser
import pythoncom
import os
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import smtplib
```

```python
engine=pyttsx3.init('sapi5')
voices=engine.getProperty('voices')
engine.setProperty('voice',voices[0].id)
```

```python
send_email_to={'m i h i r':'msonii2000@.com','h i m a n s h u':'hlade93@gmail.com','h e m a n t':'hemant110800@gmail.com','j a y a n t 99':'jayantgawali99@gmail.com','j a y a n t 98':'jayantgawali98@gmail.com','j a y a n t 100':'jayantgawali100@gmail.com','b a d s h a h':'badshahandhoneysingh100@gmail.com'}

def speak(audio):
    engine.say(audio)
    engine.runAndWait()
def findname(string):
    s=[]
    for i in string:
        if i.isalpha() or i==" ":
            s.append(i)
        else:
            break
    s=''.join(s)
    return s
def takeCommand(show):
    r=sr.Recognizer()
    with sr.Microphone() as source:
        if show:
            print("Listning...")
        r.pause_threshold=1
        audio=r.listen(source)
    try:
        if show:
            print("Recognizing...")
        query=r.recognize_google(audio,language='en-in')
        if show:
            print(f"User said: {query}\n")
    except:
        if show:
            print("Say that again please...")
        return ("None")
    return query
def wishme():
    hour=int(datetime.datetime.now().hour)
    if hour>=0 and hour<=12:
        speak("Good Morning")
    elif hour>12 and hour<=17:
        speak("Good Afternoon")
    else:
        speak("Good Evening")
    
def sendEmail(to,content):
    print("Sending Email....")
    server=smtplib.SMTP('smtp.gmail.com',587)
    server.ehlo()
    server.starttls()
    speak('please enter your email and password')
    email=input()
    password=input()
    server.login(email,password)
    speak("Login successful")
    server.sendmail(email,to,content)
    server.close()
```
